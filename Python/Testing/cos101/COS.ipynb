```python
plt.style.use('Solarize_Light2')
font = {'family' : 'Arial', 
        'weight' : 'normal', 
        'size' : 11, 
        'color' : '#818696'}

colors = ['salmon' if i <= 96 else 'red' for i in list(top_10['Progress mark(%)'])]

plt.figure(figsize=(15, 6), dpi=500)

plt.title('Top 10 students in COS101 (Progress not official)', fontdict=font)
plt.barh(y=top_10['Student number'], width=top_10['Progress mark(%)'], color=colors)

plt.xticks(np.arange(90, 101, 0.5), rotation=45)
plt.yticks(range(len(top_10)), names_top_10, fontdict=font)

plt.xlabel('Percentage')
plt.gca().invert_yaxis()

plt.xlim([90, 100])
plt.legend()
plt.show()
```
