```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df = pd.read_csv('games.csv')
```

```python
df.shape
```

```python
df["GAME_DATE_EST"] = pd.to_datetime(df["GAME_DATE_EST"])
```

```python
df.drop(['GAME_ID', 'GAME_STATUS_TEXT', 'HOME_TEAM_ID', 'VISITOR_TEAM_ID', 'TEAM_ID_home', 'TEAM_ID_away'], axis='columns', inplace=True)
```

```python
plt.figure(figsize=(10, 5), dpi=400)

df.groupby(['SEASON']).size().plot(kind='barh')

plt.title('Number of data entries for each year')
plt.ylabel('Year')
plt.show()
```

```python
df = df[df['SEASON'].isin([x + 2005 for x in range(6)])]
```
